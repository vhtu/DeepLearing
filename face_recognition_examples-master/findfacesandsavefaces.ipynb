```
from PIL import Image, ImageDraw
import face_recognition
image = face_recognition.load_image_file('/content/gdrive/My Drive/app/face_recognition_examples-master/img/groups/team2.jpg')
face_locations = face_recognition.face_locations(image)

# Array of coords of each face
# print(face_locations)

print(f'There are {len(face_locations)} people in this image')

```

```
from PIL import Image

i=1
for face_location in face_locations:
    top, right, bottom, left = face_location

    face_image = image[top:bottom, left:right]
    pil_image = Image.fromarray(face_image)
    # pil_image.show()
    pil_image.save(f'/content/gdrive/My Drive/app/face_recognition_examples-master/img/results/{i}.jpg')
    i=i+1

    
```
